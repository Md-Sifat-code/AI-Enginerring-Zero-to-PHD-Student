### This Notebook is a Visual Repsentation of 2D Sigmoid classification

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Deffine the sigmoid funtion
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
# Define points and their classes
x_points = np.array([-2,-1,1,2])
y_labels = np.array([0,0,1,1]) # 0 for the negative class and for the positive class we will see 1
```

```python
# Define values for theta and theta0
parameters = [
    (1,0),
    (-1,0),
    (0.5,0),
    (5,0)
]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

for ax, (theta, theta0) in zip(axs.ravel(), parameters):
    x_range = np.linspace(-3, 3, 300)
    z = theta * x_range + theta0
    y = sigmoid(z)

    # Plot the sigmoid curve
    ax.plot(x_range, y, label=f"Theta = {theta}, Theta0 = {theta0}")

    # Plot the data points
    ax.scatter(x_points[:2], [0, 0], color='red', label='class -1')   # Class -1 at y=0
    ax.scatter(x_points[2:], [0, 0], color='blue', label='class +1')  # Class +1 at y=0

    # Plot formatting
    ax.axhline(y=0.5, color='gray', linestyle='--')  # Decision threshold
    ax.set_ylim(-0.1, 1.1)
    ax.legend()

# Set overall figure title
fig.suptitle("Sigmoid Function for Different Theta and Theta0 Values")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
